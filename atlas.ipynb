```python
import pandas as pd
import numpy as np

YEAR = "2022"
QUARTER = "2"
vera_2022_q2 = vera[(vera["year"] == YEAR) & (vera["quarter"] == QUARTER)].copy()

# id string
id_cols = [
    "year", "quarter", "state_abbr", "state_code", "state_fips",
    "county_name", "county_code", "fips", "urbanicity", "region",
    "division", "metro_area_2000", "commuting_zone_2000"
]
# standardize FIPS
if "fips" in vera_2022_q2.columns:
    vera_2022_q2["fips"] = vera_2022_q2["fips"].astype(str).str.zfill(5)

num_cols = [c for c in vera_2022_q2.columns if c not in id_cols]
vera_2022_q2[num_cols] = vera_2022_q2[num_cols].apply(pd.to_numeric, errors="coerce")
```

```python
# check coverage of pretrial count vs rate by quarter
v22 = vera[vera["year"] == "2022"].copy()
tbl = (v22
       .groupby("quarter")
       .agg(
           rows=("fips","count"),
           nonmissing_pretrial=("total_jail_pretrial", lambda s: int(s.notna().sum())),
           nonmissing_pretrial_rate=("total_jail_pretrial_rate", lambda s: int(s.notna().sum())),
       )
       .assign(
           pct_pretrial=lambda d: d["nonmissing_pretrial"]/d["rows"],
           pct_pretrial_rate=lambda d: d["nonmissing_pretrial_rate"]/d["rows"],
       )
       .sort_index())
print(tbl)
```

```python
# make new variables!
# total_pop as denominator as requested

den_col = "total_pop" if "total_pop" in vera_2022_q2.columns else None

def safe_div(n, d):
    n = pd.to_numeric(n, errors="coerce")
    d = pd.to_numeric(d, errors="coerce")
    return np.where((d.notna()) & (d != 0), n / d, np.nan)

# Rates won't error; rows with missing/zero denom become NaN
vera_2022_q2["pretrial_rate_100k"] = (
    safe_div(vera_2022_q2["total_jail_pretrial"], vera_2022_q2[den_col]) * 1e5
    if den_col else np.nan
)

vera_2022_q2["pretrial_share"] = safe_div(
    vera_2022_q2["total_jail_pretrial"], vera_2022_q2["total_jail_pop"]
)
# keep everything (all jail/prison/admissions/race fields) for merge later
vera_keep = vera_2022_q2.copy()
```

#ATLAS

```python
import pandas as pd
atlas = pd.read_csv("/content/Food Access Research Atlas.csv")

print(atlas.shape)
print(atlas.columns.tolist())
atlas.head()
```

```python
print()
for i, col in enumerate(atlas.columns, 1):
    print(f"{i:3}. {col}")
```

```python
#county fips

def tract_to_county_fips(x):
    s = "".join(ch for ch in str(x) if ch.isdigit())
    s = s.zfill(11)[:5]                                # pad to 11, take first 5
    return s

atlas["fips"] = atlas["CensusTract"].apply(tract_to_county_fips)

# quick check～
print("county_fips nulls:", atlas["fips"].isna().sum())
print("example county_fips:", atlas["fips"].head().tolist())
```

```python
use_cols = ["Pop2010", "LAPOP1_10", "LALOWI1_10"]
atlas[use_cols] = atlas[use_cols].apply(pd.to_numeric, errors="coerce")

print(atlas[use_cols].isna().sum())
print("Pop2010 min/max:", atlas["Pop2010"].min(), atlas["Pop2010"].max())
```

```python
# was that normal? fill missing numerators with 0 (conservative)
atlas["LAPOP1_10"]   = atlas["LAPOP1_10"].fillna(0)
atlas["LALOWI1_10"]  = atlas["LALOWI1_10"].fillna(0)
```

```python
# county aggregation
g = atlas.groupby("fips", as_index=False)

county = g.agg(
    pop_total        = ("Pop2010", "sum"),
    pop_beyond_1_10  = ("LAPOP1_10", "sum"),
    lowi_beyond_1_10 = ("LALOWI1_10", "sum"),

)


# Now it's the time!!!
county["pct_distance_1_10"] = county["pop_beyond_1_10"] / county["pop_total"]
county["pct_LILA_1_10"]     = county["lowi_beyond_1_10"] / county["pop_total"]

county.head(11)
```

```python
# ~bounds check~
for col in ["pct_distance_1_10", "pct_LILA_1_10"]:
    bad = (~county[col].between(0, 1)) & county[col].notna()
    print(f"{col}: {bad.sum()} values outside [0,1]")
```

```python
import os
OUT_PATH = "/content/usda_atlas_county.csv"
county.to_csv(OUT_PATH, index=False)
print("Saved:", OUT_PATH)
```

```python
# to keep all the traces, merge county-level summary back onto tract-level data
atlas_full = atlas.merge(
    county[[
        "fips",
        "pop_total", "pop_beyond_1_10", "lowi_beyond_1_10",
        "pct_distance_1_10", "pct_LILA_1_10"
    ]],
    on="fips",
    how="left"
)

print("Shape before merge:", atlas.shape)
print("Shape after merge :", atlas_full.shape)

# peek to confirm
atlas_full[[
    "CensusTract", "fips", "Pop2010",
    "LAPOP1_10", "LALOWI1_10",
    "pop_total", "pop_beyond_1_10", "lowi_beyond_1_10",
    "pct_distance_1_10", "pct_LILA_1_10"
]].head(5)
```

```python
import os
OUT_PATH = "/content/usda_atlas_tract_and_county_summary.csv"
atlas_full.to_csv(OUT_PATH, index=False)
print("Saved:", OUT_PATH)
```
