#ATLAS

```python
import pandas as pd
atlas = pd.read_csv("/content/Food Access Research Atlas.csv")

print(atlas.shape)
print(atlas.columns.tolist())
atlas.head()
```

```python
print()
for i, col in enumerate(atlas.columns, 1):
    print(f"{i:3}. {col}")
```

```python
#county fips

def tract_to_county_fips(x):
    s = "".join(ch for ch in str(x) if ch.isdigit())
    s = s.zfill(11)[:5]                                # pad to 11, take first 5
    return s

atlas["fips"] = atlas["CensusTract"].apply(tract_to_county_fips)

# quick check～
print("county_fips nulls:", atlas["fips"].isna().sum())
print("example county_fips:", atlas["fips"].head().tolist())
```

```python
use_cols = ["Pop2010", "LAPOP1_10", "LALOWI1_10"]
atlas[use_cols] = atlas[use_cols].apply(pd.to_numeric, errors="coerce")

print(atlas[use_cols].isna().sum())
print("Pop2010 min/max:", atlas["Pop2010"].min(), atlas["Pop2010"].max())
```

```python
# was that normal? fill missing numerators with 0 (conservative)
atlas["LAPOP1_10"]   = atlas["LAPOP1_10"].fillna(0)
atlas["LALOWI1_10"]  = atlas["LALOWI1_10"].fillna(0)
```

```python
# county aggregation
g = atlas.groupby("fips", as_index=False)

county = g.agg(
    pop_total        = ("Pop2010", "sum"),
    pop_beyond_1_10  = ("LAPOP1_10", "sum"),
    lowi_beyond_1_10 = ("LALOWI1_10", "sum"),

)


# Now it's the time!!!
county["pct_distance_1_10"] = county["pop_beyond_1_10"] / county["pop_total"]
county["pct_LILA_1_10"]     = county["lowi_beyond_1_10"] / county["pop_total"]

county.head(11)
```

```python
# ~bounds check~
for col in ["pct_distance_1_10", "pct_LILA_1_10"]:
    bad = (~county[col].between(0, 1)) & county[col].notna()
    print(f"{col}: {bad.sum()} values outside [0,1]")
```

```python
import os
OUT_PATH = "/content/usda_atlas_county.csv"
county.to_csv(OUT_PATH, index=False)
print("Saved:", OUT_PATH)
```

```python
# to keep all the traces, merge county-level summary back onto tract-level data
atlas_full = atlas.merge(
    county[[
        "fips",
        "pop_total", "pop_beyond_1_10", "lowi_beyond_1_10",
        "pct_distance_1_10", "pct_LILA_1_10"
    ]],
    on="fips",
    how="left"
)

print("Shape before merge:", atlas.shape)
print("Shape after merge :", atlas_full.shape)

# peek to confirm
atlas_full[[
    "CensusTract", "fips", "Pop2010",
    "LAPOP1_10", "LALOWI1_10",
    "pop_total", "pop_beyond_1_10", "lowi_beyond_1_10",
    "pct_distance_1_10", "pct_LILA_1_10"
]].head(5)
```

```python
import os
OUT_PATH = "/content/usda_atlas_tract_and_county_summary.csv"
atlas_full.to_csv(OUT_PATH, index=False)
print("Saved:", OUT_PATH)
```
